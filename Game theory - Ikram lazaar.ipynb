# Game theory
   
   

This game theory is about the salary negotiation between a candidate and an employer.

The candidate is currently a financial auditor in one of the big four and is interviewing for the position of a financial and administrative manager in an international company in retail industry. 

He can be a great addition to the team and give them insights that they never had before which can enhance the performance of the team.

After taking the interview, he must decide whether to accept the job offer or not. 

The salary range could be set between 70-100k and the candidate is getting paid 72 K$ annually.

In this case, there are 4 senarios that both the boss and the candidate has to choose from:

1.  The company is willing to cooperate and offers 90k and the candidate cooperates by accepting the offer and feeling happy

2. The company is willing to offer 80k without negotiation (not cooperate) but the employee feels undervalued yet cooperates and accepts the offer

3. The company is willing to negociate (cooperate) by offering a salary of 90k and the employee still won't cooperate and rejects the offer.

4. Neither the company, nor the candidate are willing to cooperate, so company offers 80k and candidate rejects the offer

```python
%%time 
!pip install tabulate 
#This module will help us print a table in python
from tabulate import tabulate
import tkinter 
# this module is used to create dropdown lists in python
from tkinter import ttk
from tkinter import *
import numpy as np # numy is a library in python that allows the user to work with arrays. we will need it to create bord game
import random


# configuration of the dropdown box
root = tkinter.Tk()
root.title("Do you want to play a Game Theory Situation? ") # asking the player if he wants to play
root.geometry("500x400")

# creating the two list of options for the user
OPTION = [
    "Choose an option",
    "Yes. I want to play",
    "No. I want to exit"
    ] 

POSITION = [
    
    "I want to be the candidate", 
    "I want to be the company"
    ]
# create a pick_position function 
def pick_position(e):
    if my_combo.get() == "Yes. I want to play":
        POSITION_combo.config(value= POSITION)
    elif my_combo.get() == "No. I want to exit":
        POSITION_combo.config(value= "")

# create a dropdown box

my_combo = ttk.Combobox(root, value= OPTION)
my_combo.current(0)
my_combo.pack(pady=20)

# bind the combobox
my_combo.bind("<<ComboboxSelected>>", pick_position)

# Create POSITION combo box
POSITION_combo = ttk.Combobox(root, value= [" "] )
POSITION_combo.current(0)
POSITION_combo.pack(pady=20)


# defining a reusable variable that will allow us to link the choice of the player to the text that welcomes the player to the game
def resultout():
    myLabel = Label(root, text=clicked.get()).pack()

# If user want to play, then welcome him/her to the game
    if my_combo.get() == "Yes. I want to play" :

            user_name = input("Hi, and welcome to this game! You will play against me! Whats your name?").capitalize()
            print(f"\n{user_name} you are facing diffcult situation. Choose wisely and make the right decision!")

# Board Game
    print(f"""
        instruction: {user_name}, this is you situation, think well before answering.
        \n""")

    table = [['Company \ Candidate', """Cooperate (Accept offer)""", """Not cooperate (Dismiss offer)"""], 
             ["""Cooperate (Negociate)""", """Company gives 90k
Candidate feels happy""", """Company gives 90k
Candidate feels underevaluated"""], 
             ["""Not cooperate (No negociation)""", """Company gives 80k
Candidate feels underevaluated 
but takes the offer""", """Company gives 80k
Candidate feels underevaluated 
and refuses the offer"""]]
    print(tabulate(table, headers='firstrow', tablefmt='fancy_grid')) # as explained before, tabulate is used to create tables in python


clicked = StringVar(root) # configuration of dropdown box (continued)
myButton= Button(root,text='Result', command=resultout).pack() #this line will create a button "result". 
                                 # the choice of the player won't be confirmed unless he/she clicks on "result"
root.mainloop() # End of configuration of dropdown box

# Don't forget to close dialogue box after intering name. Otherwise, the code cell will keep running

# Create a board game with the pay-off matrix.

board_game = [[90, 0], [80, 0]]

# Save board points and results in a dictionary. C: Cooperate, NC: Not cooperate
while True:
    dictionary_results= {"(C, C)": (board_game[0][0], board_game[0][0]),
                              "(C, NC)": (board_game[0][1], board_game[1][0]),
                              "(NC, C)": (board_game[1][0], board_game[0][1]),
                              "(NC, NC)": (board_game[1][1], board_game[1][1])}
 # prompt the user to select what to do
    user_choose = input("""
    What would you do? Would you cooperate or not cooporate with the me?
    Select:
    - 'C' to Cooperate
    - 'NC' to Not cooperate
    """).upper()

# Computer (Python) chosee randomly either to cooperate or to not cooperate
    computer_choose = random.randint(0,1) # 0: cooperate; 1: not cooperate

# conditional structure with the outputs of each scenario
    if user_choose == "C" and computer_choose == 0:
            user_result = dictionary_results["(C, C)"][1]
            computer_result = dictionary_results["(C, C)"][0]
            print(f"""
        You've decided to accept the offer , Congrats for the new position!
        You negociated for {user_result} k$ annually, and I a willing to pay you {computer_result} """)

    elif user_choose == "NC" and computer_choose == 0:
            user_result = dictionary_results["(NC, C)"][1]
            computer_result = dictionary_results["(NC, C)"][0]
            print(f"""
        You've decided to not cooperate!
        you decided to walk away with {user_result} $. Not a smart choice!
        I was willing to offer you {computer_result}""")

    elif user_choose == "C" and computer_choose == 1:
            user_result = dictionary_results["(C, NC)"][1]
            computer_result = dictionary_results["(C, NC)"][0]
            print(f"""
        Even though I don't want to cooperate, you still accepted the offer.
        You will get paid {user_result} k$ annually. 
        Good job! start with a moderate salary and make your salary grow with time""")

    elif user_choose == "NC" and computer_choose == 1:
            user_result = dictionary_results["(NC, NC)"][1]
            computer_result = dictionary_results["(NC, NC)"][0]
            print(f"""
        We both agreed to disagree!
        You will walk away with {user_result} K$ annually. You feel underevaluated and unhappy because
        I was willing to pay you {computer_result}K$ annually""")

# Declaring a winner.

    if user_result > computer_result:
            print(f"""Congrats, you won!""")

    elif user_result == computer_result:
            print(f"""Well we tie!""")

    elif user_result <= computer_result:
            print(f"""Congrats to me, I beat you!""")

    play_again = input("Play again? (y/n): ")
    if play_again.lower() != "y":
        break
   
```
